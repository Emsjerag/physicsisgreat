# Theory

Suppose one has a sequence of wells, each spaced a distance $b$ apart

```python
x = np.linspace(-5,5,1000)
y = np.sum([-np.exp(-(x-a)**2 / 0.01) for a in range(-4,5)], axis=0)
plt.figure(figsize=(8,2))
plt.plot(x,y)
plt.ylabel('$V(x)$')
plt.xlabel('$x$')
plt.show()
```

This can be used as a model for a solid, where each well represents a nucleus or proton. If one inserts an electron into this system, what are the allowed energy eigenvalues/eigenvectors of **bound** energies? (Remember that there are only finitely many bound energies for a finite quantum well).

**LINEAR COMBINATION OF ATOMIC ORBIRTALS (LCAO) Approximation**: This model **assumes** that molecular states are a linear combination of known atomic states. We'll use the notation

$$\ket{i}_j$$

where this represents the $i$th excited state of the $j$th well. The LCAO approximation says that any state can be represented as 

$$\ket{\psi} = \sum_{j} \sum_i \ket{i}_j$$

Thus the $\ket{i}_j$ form a complete basis. Suppose we have 3 wells, each with two states (a ground and excited state). In vector notation, we can write any state as

$$\ket{\psi} = \begin{bmatrix}c_1\\ c_2\\ c_3\\ c_4\\ c_5 \\ c_6  \end{bmatrix} \begin{matrix} \to \ket{0}_0\\ \to \ket{1}_0\\ \to \ket{0}_1\\ \to \ket{1}_1\\ \to \ket{0}_2\\ \to \ket{1}_2\\ \end{matrix}$$

where the $c$'s are the component of each of the $\ket{i}_j$s of the basis:

$$\ket{\psi} = c_1\ket{0}_0 + c_2\ket{1}_0 + c_3\ket{0}_1 + c_4\ket{1}_1 + c_1\ket{0}_2 + c_1\ket{1}_2$$

The Hamiltonian $H = T+\sum_i V_i$ of the full well system can be written in the basis of these $\ket{i}_j$s where each element of the hamiltonian matrix is

$$\bra{i'}_{j'}H\ket{i}_j$$

once we have this matrix, we can solve for the eigenvalues and eigenvectors and we're done. We just need a way to get these elements. For this we use

$$H = H_0 + V'$$

where $H_0 = T+V_0$ is the Hamiltonian of a single well and $\sum_{i \neq 0} V_i$ is the potential from all the adjacent wells. Then we have two cases

**Case 1**: $j=j'$ (considering overlapping wavefunctions from the same well)

$$\begin{align*} \bra{i'}_{j}H\ket{i}_j &= \bra{i'}_{j}H_0\ket{i}_j + \bra{i'}_{j}V'\ket{i}_j \\ &= E_i \delta_{i,i'} + \int_{-\infty}^{\infty} \phi_i^{*}(x) V'(x)\phi_{i'}(x) dx \end{align*}$$

where $\delta_{i,i'}=1$ when $i=i'$ and 0 otherwise. We know the $E_i$'s and $\phi_i$'s from solving everything for a single well, and we know $V(x)$ (the shape of the well) so this can easily be numerically solved for.

**Case 2**: $j \neq j'$ (considering overlapping wave functions from different wells). In this case, only consider $j' = j \pm 1$ (adjacent wells) since wavefunctions from far away wells don't overlap

$$\begin{align*} \bra{i'}_{j}H\ket{i}_{j \pm 1} &= \bra{i'}_{j}H_0\ket{i}_{j \pm 1} + \bra{i'}_{j}V'\ket{i}_{j \pm 1} \\ &= \int_{-\infty}^{\infty} \phi_i^{*}(x) V'(x)\phi_{i'}(x \pm b) dx \end{align*}$$

The first term $\bra{i'}_{j}H_0\ket{i}_{j \pm 1}$ is neglected since the overlap of these wavefunctions is especially small:

* $V=0$ in the overlap region of these wavefunctions
* $T\ket{i}_{j \pm 1}$, which is essentially a second derivative of the wave function, is **extremely** small in the overlap region

Now that we know how to compute all the matrix elements of the Hamiltonian, we can obtain it, find the energy eigenvalues, and we're done!

```python
def solve_single_well(x, V, V0):
    N = len(x)
    dx = np.diff(x)[0]
    main_diag = 2*np.ones(N)/dx**2 + V(x,V0)
    off_diag =  -np.ones(N-1)/dx**2
    Es, psis = eigh_tridiagonal(main_diag, off_diag, select='v', select_range=(-V0,0))
    psis = psis/np.sqrt((np.abs(psis)**2).sum(axis=0) * dx)
    return Es, psis.T
```

# Example: Finite Square Well

* $V_0 = 1$eV
* Well Width $2a=0.35$nm
* Well spacing $b = 3 \cdot 2a$

```python
V0 = 5 #eV
hbar = 6.582112e-16 #eV s
m = 0.510999e6 / (3e8)**2 #eV s^2 / m^2
a = 0.35e-9 / 2 # m 
f = hbar**2 / (2*m*a**2)
```

```python
N = 1000000
V0 = V0 / f
b = 3 * 2
x = np.linspace(-25,25,N)
def V1(x, V0):
    return -V0*np.array((x>=-1)*(x<=1), dtype=float)
def V2(x,V0):
    return -V0*np.exp(-x**2/0.1)
```

```python
plt.figure(figsize=(6,3))
plt.plot(x, V1(x,V0))
plt.show()
```

```python
Es, psis = solve_single_well(x, V1, V0)
```

Look at the energies

```python
f*Es
```

Plot all the wavefunctions

```python
plt.figure(figsize=(6,3))
[plt.plot(x, psi) for psi in psis]
plt.show()
```

We need to be able to evaluate $\int_{-\infty}^{\infty} \phi_i^{*}(x) V'(x)\phi_{i'}(x \pm b) dx$. For this we need a function that can compute the wavefunctions at any point $x$

```python
psis_f = [interp1d(x, psi, bounds_error=False, fill_value=(0,0)) for psi in psis]
```

Define $V'(x)$: the potential from all the wells, **except** the one centered at $x=0$.

```python
def Vp_f(x, b, V0):
    return V1((x+b/2)%b-b/2,V0) - V1(x,V0)
```

Plot the ground state wavefunctions from consecutive wells, along with $V'(x)$

```python
x_large = np.linspace(-10,10,10000)
plt.figure(figsize=(6,3))
plt.plot(x_large, psis_f[0](x_large)**2)
plt.plot(x_large, psis_f[0](x_large - b)**2)
plt.plot(x_large, Vp_f(x_large,b,V0)/V0, color='red')
plt.show()
```

Function that computes $\int_{-\infty}^{\infty} \phi_i^{*}(x) V'(x)\phi_{i'}(x \pm b) dx$

```python
def compute_integral(psis_f, i1, i2, b, V0, n):
    return quad(lambda x: psis_f[i1](x)*psis_f[i2](x-n*b)*Vp_f(x, b, V0),
                -20, 20, points=[n*b for n in range(-6,6)], limit=1000)[0]   
```

```python
f*compute_integral(psis_f,0,0, b=6, V0=V0, n=1)
```

## Fill the Hamiltonian matrix $H$

**Part 1**: Create Matrix

```python
N_wells = 10
N_states = len(Es)
H = np.zeros((N_wells*N_states,N_wells*N_states))
```

**Part 2**: Fill Off diagonal

```python
psis_tot = N_wells*psis_f
for i1 in range(H.shape[0]):
    for i2 in range(i1, H.shape[0]):
        well1_idx = i1//N_states
        well2_idx = i2//N_states
        H[i1,i2] = compute_integral(psis_tot,i1,i2,b=b,V0=V0,n=well2_idx-well1_idx)
H += H.T
```

**Part 3**: Fill Main diagonal

```python
main_diag = np.array(list(Es)*N_wells)
H[np.diag_indices(H.shape[0])] = main_diag
```

Show a visual representation of the Hamiltonian

```python
plt.imshow(H, cmap='Greys')
plt.colorbar()
plt.show()
```

```python
eigenenergies, eigenvectors = np.linalg.eigh(H)
```

```python
plt.figure(figsize=(2,4))
plt.eventplot(f*eigenenergies, orientation='vertical', colors='r', linewidths=0.5)
plt.show()
```
